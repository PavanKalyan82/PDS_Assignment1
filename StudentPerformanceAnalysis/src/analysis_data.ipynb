```python
import pandas as pd
```

```python
df = pd.read_csv('/content/clean_data/StudentPerformance_clean.csv')
df=df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df["Percentage"]=(df["math score"]+df["reading score"]+df["writing score"])/3
print(df["Percentage"])
```

```python
df['Percentage'] = (df['math score']+df['reading score']+df['writing score'])/3
print(df["Percentage"])
```

```python
import seaborn as sns
import numpy as np
```

```python
custom_palette = ["red", "blue"]
sns.barplot(x ='gender', y ='Percentage', data = df,
            palette =custom_palette, estimator = np.std)
```

```python
df['gender'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
labels=['Female', 'Male']

plt.pie(df['gender'].value_counts(),labels=labels,explode=[0.1,0.1],
        autopct='%1.2f%%',colors=['#FFFF00', '#0000FF'], startangle=90)
```

```python
ax=sns.distplot(df['Percentage'], kde = False, color ='red', bins = 30)
fig=ax.get_figure()
```

```python
custom_palette = ["green", "blue"]
ax=sns.jointplot(x ='Percentage', y ='math score',hue="gender", data = df,palette=custom_palette)
```
